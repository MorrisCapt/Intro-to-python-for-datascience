1.Using OpenCV (cv2)

```python
#Ensure your image is in the same directory as your script or provide the full file path to the image.
import cv2  
import matplotlib.pyplot as plt  

# Load the image  
image = cv2.imread("jade.jpg")  

# Convert from BGR to RGB (OpenCV loads images in BGR format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

# Display the image  
plt.imshow(image)  
plt.axis("off")  # Hide axes
plt.show()  
```

**Understand the difference between loading (json) and reading (CSV )**
Both JSON and CSV are widely used formats for storing and exchanging data, but they have different structures and are handled differently in Python.

```python
#Loading a JSON file
import json

with open("data.json", "r") as f:
    data = json.load(f)  # Loads JSON into a Python dictionary
print(data)
```

```python
#Reading a CSV file
import pandas as pd

df = pd.read_csv("data.csv")  # Reads CSV into a Pandas DataFrame
print(df.head())
```

**Key Takeaways**
1. JSON is hierarchical, while CSV is tabular.
2. Use json.load() when dealing with structured data like API responses.
3. Use pd.read_csv() when working with tabular datasets.

**The difference between writing (CSV) and dumping (json)**
Both CSV writing (to_csv()) and JSON dumping (json.dump()) are used to save data, but they work differently based on the data format.

```python
#Writing to CSV
import pandas as pd

data = {"Name": ["Alice", "Bob"], "Age": [25, 30]}
df = pd.DataFrame(data)

# Save to a CSV file
df.to_csv("data.csv", index=False)
```

```python
#Dumping to JSON
import json

data = {"Name": "Alice", "Age": 25, "Hobbies": ["Reading", "Cycling"]}

# Save to a JSON file
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)
```

**Key Takeaways**
1. CSV (to_csv()) is best for structured, tabular data.
2. JSON (json.dump()) is better for nested, hierarchical data.
3. CSV is easier for spreadsheets and databases, while JSON is common for APIs and configurations.
